```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'


from skimage import data
from matplotlib import pyplot as plt

import cv2
```

```python
# Load image and get its dimensions
image = data.astronaut()
plt.imshow(image)
rows, cols, dim = image.shape
print(rows, cols, dim)

gray= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)


```

```python
# sift.detect() function finds the keypoint in the images.
#  Each keypoint is a special structure which has many attributes 
#  It has its (x,y) coordinates, 
#             size of the meaningful neighbourhood, 
#             angle which specifies its orientation, 
```

```python
sift = cv2.SIFT()
kp = sift.detect(gray,None)

img = cv2.drawKeypoints(gray,kp)

img2 = img[:,:,::-1]
plt.imshow(img2)
```

```python
#  OpenCV also provides cv2.drawKeyPoints() function 
#  which draws the small circles on the locations of keypoints. 
#  If you pass a flag, cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS to it, 
#  it will draw a circle with size of keypoint and it will even show its orientation. See below example.
```

```python
img = cv2.drawKeypoints(gray,kp,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
img3 = img[:,:,::-1]
plt.imshow(img3)
```
