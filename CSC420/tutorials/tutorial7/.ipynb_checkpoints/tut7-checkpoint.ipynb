```python
for det in detection['dets'][0][0]:
    xleft, ytop, xright, ybottom, _, _ = det
    print(det)
    cv2.rectangle(imgLeft, (int(xleft), int(ytop)),
                  (int(xright), int(ybottom)), [255,255,0],3)

    plt.imshow(imgLeft[:,:,::-1])
    plt.show()
```

```python
gt_mask = cv2.imread('./data/gt_left/004945.png', cv2.CV_LOAD_IMAGE_GRAYSCALE)
plt.imshow(gt_mask)
plt.show()
gt_mask.shape
print(gt_mask.max())
print(gt_mask.min())
# 16 is background
obj_ids = np.unique(gt_mask)
print(obj_ids)
number_object = obj_ids.shape[0]

# norm ids
count = 0
for o_id in obj_ids:
    gt_mask[gt_mask == o_id] = count
    count += 1
```

```python
import matplotlib.colors as mcolors
import webcolors

base_COLORS = []

for key, value in mcolors.CSS4_COLORS.items():
    rgb = webcolors.hex_to_rgb(value)
    base_COLORS.append([rgb.blue, rgb.green, rgb.red])
base_COLORS = np.array(base_COLORS)

np.random.seed(99)
base_COLORS = np.random.permutation(base_COLORS)

colour_id = np.array([(id) % len(base_COLORS) for id in range(number_object)])

COLORS = base_COLORS[colour_id]
COLORS = np.vstack([[0, 0, 0], COLORS]).astype("uint8")
mask = COLORS[gt_mask]

imgLeft = cv2.imread('./data/left/004945.jpg')
output = ((0.4 * imgLeft) + (0.6 * mask)).astype("uint8")
plt.imshow(output[:,:,::-1])
```
